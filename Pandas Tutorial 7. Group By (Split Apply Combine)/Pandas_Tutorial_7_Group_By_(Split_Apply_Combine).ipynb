```python
import pandas as pd
```

```python
# In this tutorial we'll be looking at Pandas groupby feaure
# We have excel sheet which contains weather of 3 different cities
# Ney York, Mumbai, Paris
```

```python
# I'll group each cities dataframe and perform an analysis
```

```python
df = pd.read_csv("weather_by_cities.csv")
df
```

```python
g = df.groupby('city')
g
```

group_by_cities.png

```python
# g looks like the picture above
```

```python
# How to access the each groups
# We'll have to make an iterator
for city, city_df in g:
  print(city)
  print(city_df)
```

```python
# Acces specific groupby
g.get_group('mumbai')
```

```python
# Get maximum of temperature in each city and all the columns
g.max()
```

```python
# This is really interesting because now what you're doing is, you're first divinding your dataset into different groups based on cities and then you're running analytics
# on each of these groups and then you're combining the results into this single dataframe so this process of dividing your data into multiple groups and then applying
# some analytics to get aggregated result is caleed "SplitApplyCombine"
# Refer to the picture below
```

```python
# What is the average windspeed
g.mean()
```

```python
# Get all the analytics in one shot
g.describe()
```
