```python
import pandas as pd
```

```python
# first way of creating a dataframe (using CSV)
df = pd.read_csv("weather_data.csv")
df
```

```python
# Second way of creating Dataframe (using EXCEL)
df = pd.read_excel('weather_data.xlsx', 'Sheet1')
df
```

```python
# Third way of creating a DataFrame (Using a Python Dictionary)
weather_data = {
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'windspeed': [6,7,2],
    'event': ['Rain', 'Sunny', 'Snow']
}

df = pd.DataFrame(weather_data)
df
```

```python
# fourth of way of creating a DataFrame is (using a tuples list)
# note that these are tupules and when you make a dataframe out of tuples you need to provide columns name, because these are just rows
weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]

df = pd.DataFrame(weather_data, columns =['day', 'temperature','windspeed', 'event'])
df
```

```python
# Fifth way to make a dataframe is using list of dictionaries
# The difference between making a DataFrame from the third and fifth way is different
# Here each element in the list represents one row

weather_data = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
    
]

df = pd.DataFrame(weather_data)
df
```
